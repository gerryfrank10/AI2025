### Data Loading and Data Pre-processing


```python
#Importing the required libraries
import pandas as pd
import numpy as npy

# loading the dataset
Churn_data= pd.read_csv("E:/Sujith/Churn Prediction/BankChurners.csv")

Churn_data.head(5)
```

```python
Churn_data.shape
```

```python
Churn_data.columns
```

```python
Churn_data.info()
```

```python
# Drop duplicate rows if any
Churn_data.drop_duplicates(inplace=True)
```

```python
Churn_data1 = Churn_data.rename(columns={
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1': 'NB_Classifier_1',
    'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2': 'NB_Classifier_2'
})
print(Churn_data1)
```

```python
null_values = Churn_data1.isnull().sum()
null_values
```

```python
# Convert categorical columns to category dtype
category_columns = ['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category', 'Attrition_Flag']
for col in category_columns:
    Churn_data1[col] = Churn_data[col].astype('category')

category_columns
```

```python
Churn_data1.describe()
```

### Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))

sns.histplot(Churn_data1["Customer_Age"],kde=True,color="green")
plt.title('Age Distribution for Customers')
plt.xlabel('Age')
plt.ylabel('Density')

plt.show()
```

```python
print("Education Level Frequency:\n", Churn_data1['Education_Level'].value_counts())
```

```python
edu_attrition = pd.crosstab(Churn_data1['Attrition_Flag'], Churn_data['Education_Level'])
print("\nEducation vs Attrition:\n", edu_attrition)

# Visualization
plt.figure(figsize=(10, 5))
sns.countplot(data=Churn_data1, x='Education_Level', hue='Attrition_Flag', palette='pastel')
plt.title('Customer Attrition by Education Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Frequency count
print("Marital Status Frequency:\n", Churn_data1['Marital_Status'].value_counts())
```

```python
# Frequency count
print("Income Category Frequency:\n", Churn_data1['Income_Category'].value_counts())
```

```python
income_attrition = pd.crosstab(Churn_data['Attrition_Flag'], Churn_data['Income_Category'])
print("\nIncome Category vs Attrition:\n", income_attrition)

# Visualization
plt.figure(figsize=(10, 5))
sns.countplot(data=Churn_data, x='Income_Category', hue='Attrition_Flag', palette='Set3')
plt.title('Customer Attrition by Income Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count the frequency of each card category
card_counts = Churn_data1['Card_Category'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(card_counts, labels=card_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))

plt.title('Distribution of Card Categories')
plt.axis('equal')  # Ensures pie is a circle
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=Churn_data1, x='Card_Category', hue='Attrition_Flag', palette='pastel')

plt.title('Card Category Distribution by Attrition Flag')
plt.xlabel('Card Category')
plt.ylabel('Customer Count')
plt.tight_layout()
plt.show()
```
