```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
# iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
sns.relplot(data=iris, x='sepal_length', y='sepal_width', hue='species');
```

```python
# iris.plot(kind='scatter', x='petal_length', y='petal_width',)
# import matplotlib.pyplot as plt
# plt.title('Petal length vs Petal width')
sns.relplot(x='petal_length', y='petal_width', data=iris,hue='species');
```

```python
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
km = KMeans(n_clusters=3)
iris['cluster'] = km.fit_predict(X)
iris['cluster'] = iris['cluster'].astype('category')
px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_width', color='cluster')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = iris['species']
y = le.fit_transform(y)
iris['cluster'] = le.fit_transform(iris['cluster'])
correct_labels = sum(y == iris['cluster'])
print(f'Number of correct labels: {correct_labels} out of {len(iris)}')
```
